### Data standardization

```python
# Transforming coulumsfor ML training
import sklearn
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import make_column_transformer
```

```python
oe = OrdinalEncoder()
ct = make_column_transformer(
    (oe,["venue","team1","team2","toss_winner","toss_decision","result","eliminator","method"]),
    remainder="passthrough"
)
ct.set_output(transform = 'pandas')
ipl_df = ct.fit_transform(ipl)
```

```python
le = LabelEncoder()
ipl_df['remainder__winner'] = le.fit_transform(ipl_df['remainder__winner'])
ipl_df.head(2)
```

```python
#Droping columns
X= ipl_df.drop(['remainder__id','remainder__city','ordinalencoder__method','remainder__date','remainder__player_of_match','remainder__winner','remainder__umpire1','remainder__umpire2'],axis = 1)
Y = ipl_df['remainder__winner']
```

```python
#Using standrdization for input data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() 
X_scaled = scaler.fit_transform(X)
```

### Building a model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_scaled,Y,test_size=0.3,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100,random_state=42)
#Training the model
rf.fit(X_train,y_train)
#Prediction for the model
y_pred = rf.predict(X_test)
```

```python
#Evaluation of the model
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test,y_pred)
print("The accuracy of the model is %.3f" % acc)
```

```python
#Removing the na values
spotify = spotify.dropna()
```

```python
features = ['danceability', 'energy', 'speechiness', 'acousticness', 'instrumentalness', 'valence']
```

```python
#Removing the ' ' values
spotify[features] = spotify[features].apply(pd.to_numeric,errors='coerce')
spotify[features].dropna(inplace=True)
```

```python
#Scatter plot
plt.figure(figsize=(10, 5))
sns.lineplot(x='year', y='popularity', data=spotify, estimator='mean')
plt.title("Average Popularity Over the Years")
plt.xlabel("Year")
plt.ylabel("Average Popularity")
plt.show()
```

```python
#Boxplot for distribution of key features
plt.figure(figsize=(10, 6))
sns.boxplot(data=spotify[features])
plt.xticks(rotation=45)
plt.title("Distribution of Key Features")
plt.show()
```

```python
spotify_corr = spotify[features].corr()
```

```python
# Correlation heat map
plt.figure(figsize=(10,8))
sns.heatmap(spotify_corr,annot=True,cmap='coolwarm',linewidths=0.5)
```

```python
from sklearn.cluster import KMeans
# Clustering songs based on acoustic features
scaler = StandardScaler()
spotify_scaled = scaler.fit_transform(spotify[features])

kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(spotify_scaled)
spotify['Cluster'] = kmeans.labels_

plt.figure(figsize=(8, 6))
sns.scatterplot(x=spotify['danceability'], y=spotify['energy'], hue=spotify['Cluster'], palette='deep')
plt.title("Clustering Songs Based on Acoustic Characteristics")
plt.xlabel("Danceability")
plt.ylabel("Energy")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=spotify['liveness'], y=spotify['speechiness'], hue=spotify['Cluster'], palette='deep')
plt.title("Clustering Songs Based on Speechiness Characteristics")
plt.xlabel("liveness")
plt.ylabel("speechiness")
plt.show()
```

```python
#Splittng the data to input data and outcome
X = cancer.drop('diagnosis',axis=1)
Y = cancer['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=20)
```

```python
#Training model
from sklearn.naive_bayes import BernoulliNB
nb= BernoulliNB()

nb.fit(X_train,y_train)
#Prediction
y_pred = nb.predict(X_test)
```

```python
#Evaluating the model
from sklearn.metrics import accuracy_score,classification_report
accuracy = accuracy_score(y_test,y_pred)
classify = classification_report(y_test,y_pred)
print("The accuracy of the model is %.3f" % accuracy)
print(classify)
```
