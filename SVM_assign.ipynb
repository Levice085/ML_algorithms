```python
# Data type conversion from object to string
income_eval[["workclass","education","marital-status","occupation","relationship","race","sex","native-country","income"]] = income_eval[["workclass","education","marital-status",
                                                                                                                                           "occupation","relationship","race","sex","native-country",
                                                                                                                                          "income"]].astype(str)
                                                                                                                                           
#Conversion from object to integer
income_eval[["fnlwgt", "education-num","hours-per-week"]]=income_eval[["fnlwgt", "education-num","hours-per-week"]].astype(int)
```

### Converting the columns to data that can be readable for machine learning
#### **Use make-column-transformer**

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder()
```

```python
income_eval.head(2)
```

```python
from sklearn.compose import make_column_transformer

ct = make_column_transformer(
    (oe,['workclass','education','marital-status','occupation','relationship','race','sex']),
    remainder = 'passthrough'
    
)
ct.set_output(transform='pandas')
df_income =ct.fit_transform(income_eval)
```

```python
df_income.head()
```

```python
#Splitting data into X and y
X = df_income.iloc[:,:12]
y = df_income.iloc[:,12]
print(X.shape)
print(y.shape)
X.head()
```

### Converting the income column using label encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
income_eval['income']=le.fit_transform(income_eval['income'])
income_eval['income'].value_counts()
```

```python
#Splitting data into training and splitting datasets
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2 ,random_state=100)
```

```python
X_train.shape
y_train.shape
```

```python
#Training the model
from sklearn.svm import SVC
modelsvc = SVC()
```

```python
#Fitting th model
modelsvc.fit(X_train,y_train)
```

```python
#Predicting the model
y_pred= modelsvc.predict(X_test)
```
